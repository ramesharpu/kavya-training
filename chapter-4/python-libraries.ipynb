# Python Libraries

Python Libraries are collections of **pre-written code, modules and functions** that allow developers to perform complex tasks without writing the logic from scratch.

The ecosystem consists of two main branches:
* build-in python standard libraries
* hundres of thousands of third-party packages ==> available via the python package index (PyPI) ==> https://pypi.org/


---
# Key Build-in Libraries(Standard Library)

These modules require no installation and are ready right out of the box
**Example:**
* **os**: Handles interaction with operating system file directories
* **sys**: Modifies system-specific parameters and controls runtime script settings
* **datatime**: Essential module built for parsing, calculating and manipulating dates and times
* **math**: math library provides a comphrehensive collection of mathematical functions and constants for floating-point arithmetic
