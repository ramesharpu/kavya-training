# Local and Global Variables

### 1. Local variable
Local varible is defined inside the method

```python
def m1():
    a = 5 # Local Variable and its scope is inside m1 only 
    print(a)
    
def m2():
    print(a) # it will give error message
    
m1()
m2()
```

### 2. Global Variable
Global varible is defined outside the method

```python
a = 10 # Global variable
def m1():
    print("from m1, a = " ,a)

def m2():
    print("from m2, a = ", a)    
    
m1()
m2()

```

# Mutable and Immutable concepts

* variable **num** value is does not change as we are not updating it
* variable **x** value changes as we are performing addition operation on it

```python
def f1(x):
    print("intial value of x = ", x)
    print(id(x))
    print("---")
    x = x + 2
    print("modified value of x = ", x)
    print(id(x))

num = 5
print("before: id of num = ",id(num))
print("value of num = ", num)
f1(num)
print("after: id of num = ",id(num))
print("value of num = ", num)
```

```python
def m1(x):
    print("initial value of x = ", x)
    print("location of x = ", id(x))

    print("\n-------------------")
    print("Updating the value of x to 56")
    x = 56
    print("modified value of x = ", x)
    print("location of x = ", id(x))

y = 35
m1(y)

print("\n--------------")
print("y value = ", y)
print("y's location = ", id(y))
```
